# Fragrantica Data EDA
Explore the seed perfume dataset and community voting patterns.

```python
import json
import pandas as pd
import matplotlib.pyplot as plt

with open('../backend/data/seed_perfumes.json') as f:
    perfumes = json.load(f)

df = pd.DataFrame(perfumes)
print(f'Total perfumes: {len(df)}')
df[['name','brand','concentration','community_overall_rating']].head(20)
```

```python
# Season voting patterns
seasons = ['season_spring_votes','season_summer_votes','season_fall_votes','season_winter_votes']
df[seasons].sum().plot(kind='bar', color=['#34d399','#fbbf24','#f87171','#60a5fa'])
plt.title('Total Season Votes Across Seed Perfumes')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# Top perfumes by overall rating
df.nlargest(10, 'community_overall_rating')[['name','brand','concentration','community_overall_rating']]
```

```python
# Note frequency analysis
from collections import Counter
all_notes = []
for _, row in df.iterrows():
    all_notes.extend(row.get('top_notes', []))
    all_notes.extend(row.get('middle_notes', []))
    all_notes.extend(row.get('base_notes', []))

top_notes = Counter(all_notes).most_common(15)
notes_df = pd.DataFrame(top_notes, columns=['note','count'])
notes_df.plot(kind='barh', x='note', y='count', color='#a78bfa', legend=False)
plt.title('Most Common Notes in Seed Dataset')
plt.tight_layout()
plt.show()
```
