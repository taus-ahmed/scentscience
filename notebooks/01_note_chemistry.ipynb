# Note Chemistry Database
Validate and explore the 50-note chemistry database.

```python
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

with open('../backend/data/notes_chemistry.json') as f:
    notes = json.load(f)

df = pd.DataFrame(notes)
print(f'Total notes: {len(df)}')
df.head()
```

```python
# Family distribution
plt.figure(figsize=(12, 4))
df['family'].value_counts().plot(kind='bar', color='#a78bfa')
plt.title('Notes by Family')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Heatmap of note properties
numeric_cols = ['volatility','heat_performance','cold_performance',
                'humidity_performance','dry_performance','skin_bonding',
                'projection_strength']
fig, ax = plt.subplots(figsize=(14, 12))
sns.heatmap(
    df.set_index('name')[numeric_cols],
    cmap='RdYlGn', vmin=1, vmax=10, annot=True, fmt='.0f',
    ax=ax, cbar_kws={'shrink': 0.5}
)
plt.title('Note Chemistry Matrix')
plt.tight_layout()
plt.show()
```
